**standardaization**

-standardaization means scaling the data into one scale

- we have different cols has diff units son that the value will vary

- one column has very huge values

- another column has very less values

- so it is important to scale all type of units under one scale

- we have 2 procedures

- standardaization

    - z-score:
        
        - the value ranges from -3 to 3
        
        

-Normalization

    - min max scalar

```python
#s-1
# take prevailing wage
# z-score = x-mean/sigma


#s-2- calculate mean of prevailing wage
#s-3-calculate std of prewailing wage
#s-4- numerator:pwage-mean
#s-5-pwage_zscore=nr/dr
```

```python
pwage=visa_df['prevailing_wage']
pwage_mean=visa_df['prevailing_wage'].mean()
pwage_std=visa_df['prevailing_wage'].std()
Nr=pwage-pwage_mean
visa_df['prevailing_wage_z']=Nr/pwage_std
```

$min$-$max$-$scalar$

```python
# x-x min/(x_max-x_min)
# x=data column
# x_min
# x_max
# nr=x-x_min
# dr=x_max-x_min
# nr/dr
```

```python
pwage=visa_df['prevailing_wage']
pwage_min=visa_df['prevailing_wage'].min()
pwage_max=visa_df['prevailing_wage'].max()
nr=pwage-pwage_min
dr=pwage_max-pwage_min

visa_df['prevailing_wage_norm']=nr/dr
visa_df['prevailing_wage_norm']
```
