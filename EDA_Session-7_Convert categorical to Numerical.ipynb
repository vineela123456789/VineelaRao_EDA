**one hot encoder**

- one hot encoder name says that one will on and other will off

- for eg: case status has 2 labels 
     
        - certified
        
        - Denied   
        
- when you apply one hot encoding on case status , it creates 2 more extra column

    - case_status_Certifird
    
    - case_status_Denied
    
|case_status|case_status_certified|case_status_denied|
|-|-|-|
|Certified|1|
|Denied|0|

**Advantages**

- when you develop ML model it is very important the columns should be independent each other

- so here case status creating 2 extra columns

- which are independent each other ,

- 90 degrees phase shift

- which means perpendicular to each other 

- which means orthoganal to each other

**Disadvantages**

- if a column ha 100 unique labels , 100 new columns will be created

- the data will become sparse, which means huge

- columns are more means dimensions are more

- the processing time is more

- memory consumption is more

- Curse of Dimensionality
